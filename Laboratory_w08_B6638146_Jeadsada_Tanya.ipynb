# **K Means Cluster Creation**

```python
from sklearn.cluster import KMeans
```

```python
X = df.select_dtypes(include="number")
```

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

# **Evaluation**

```python
from sklearn.metrics import accuracy_score
true_private = df["Private"].map({"No": 0, "Yes": 1}).values
pred = kmeans.labels_
if accuracy_score(true_private, pred) < accuracy_score(true_private, 1 - pred):
    pred = 1 - pred
```

```python
df["Cluster"] = pred
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(true_private, df["Cluster"])
print(cm)
print("\nClassification Report:")
print(classification_report(true_private, df["Cluster"]))
```
